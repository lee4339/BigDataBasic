```python
n = 3 
m = 5

print(n == m)
print(n > m)
print(n <= m)
print(n != m)
```

```python
if n == m:
  print("두 값이 같습니다.")
else :
  print("두 값이 다릅니다.")

if n >= m: #if elif는 순서대로 참인 경우에서 빠져나감 참 이후의 조건은 무시
  print("n이 m보다 작거나 같습니다.")
elif n % 3 == 0:
  print("n은 3의 배수입니다.")
else :
  print("마지막 else.")
```

```python
x = [1, 2, 3]

if 3 in x:
  print("리스트에 3이 있습니다.")

if 5 not in [3, 2, 1]:
  print("리스트에 5가 없습니다.")
```

```python
n = 3
y = {1, 2, 3}

while n > 0:
  print(n)
  n = n - 1

while 3 in y:
  print('3이 세트 y에 있습니다.')
  break # y.remove(3)
```

```python
x = [1, 2, 3, 4, 5]
z = [n ** 2 for n in  x] # 인라인함수 --> n안에 x를 넣는데 n은 2의 제곱이다.

print(z)
```
